Exercise 1: Implementing an Undirected Graph with an Adjacency List Consider the implementation of an undirected graph with an adjacency list:
graph = {
'A': ['B', 'C'],
'B': ['A', 'C'], '
C': ['A', 'B']
}


```python
import networkx as nx
import matplotlib.pyplot as plt
# Create an empty undirected graph according to the adjancent list given
G = nx.Graph()
# Add nodes
G.add_node('A')
G.add_node('B')
G.add_node('C')
# Add edges
G.add_edge('B', 'C')
G.add_edge('A', 'C')
G.add_edge('A', 'B')

#1. Add node D to exisiting graph
G.add_node('D')

#2. Add new edge
G.add_edge('B',"D")

#3. Removing an edge
G.remove_edge('B','D')

#4. Traverse the graph Note that we are using DFS here, which stand for "Depth-First Search"
def dfs(G, node, visited):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbor in G[node]:
            dfs(G, neighbor, visited)
#Usage
visited = set()
dfs(G, 'A', visited)

print(G)

#5. Graph Visualization: This is an undirected graph because the edges have no direction and D is an isolated node.
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Undirected Graph")
plt.show()
```

Execise 2: Finding Shortest Paths in a Weighted Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a weighted directed graph
G = nx.DiGraph()

# then we add edges
G.add_edge('A', 'B', weight=2)
G.add_edge('A', 'C', weight=5)
G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=4)
G.add_edge('C', 'D', weight=3)

#1. Shortest Path Calculation, Using Dijkstra's algorithm
shortest_path = nx.shortest_path(G, source='A', target='D', weight='weight')
print("Shortest Path:", shortest_path)

#2. Updating edge weights: Update 'B' to 'C' to 3
if G.has_edge('B', 'C'):
    G['B']['C']['weight'] = 3


#4. Graph Visualization: Draw the graph using NetworkX
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Weighted Directed Graph")
plt.show()
```
